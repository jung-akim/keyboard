#### Imputation by Multiple imputation by chained equations

Description of MICE algorithm:<br>
https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3074241/<br><br>
Note:<br>
Re-define 'rating' variable since the missing ratings are obviously because of zero reviews for the products.<br>
'rating' is missing for all observations with zero number of ratings/reviews which makes sense, but 92% of the observations have zero reviews and 'rating' is a strong predictor for number of items sold.
Before imputation, 'rating' was an average rating of all the ratings for the product. But, if we were to imputate the 92% of the ratings, we would have to define it differently, i.e., Estimated expected rating for the product.
So, 8% of the ratings have a sample mean as their estimate of expected rating.
92%, the imputed ratings, would have the estimated expected rating by MICE algorithm which uses cycles of linear regression with all the other predictors regressing on 'rating'.

```python
imp = mice.MICEData(x_train)
imp.set_imputer('rating', formula='price + num_ratings + watcher + shipping + free_return + shipping + free_return + open_box + pre_owned + refurbished + benefits_charity')
print("Cycle")
for j in range(int(100 * np.sum(pd.isna(x_train.rating))/len(x_train))):
    imp.update_all()
    print(f"{j+1}th done.", end = ' ')
x_train = imp.data.set_index(x_train.index)
```

#### Interaction effects

```python
log_y = np.log(y_train+1); X = x_train
mreg = smf.ols(formula = 'log_y ~ X', data = X, missing = 'drop').fit()
def plot_interaction(var1, var2, title, xlim = None):
    log_y = np.log(y_train+1)
    fig, ax = plt.subplots(figsize=(20, 5))
    interaction_plot(x = var1, trace = var2, response = log_y, colors=['red','blue'], ax=ax)
    if xlim:
        plt.xlim(xlim)

    X = pd.concat([x_train, (var1 * var2)], axis =1)
    m = smf.ols(formula = 'log_y ~ X', data = X, missing = 'drop').fit()
    plt.title(title)
    plt.show()
    df = pd.concat([pd.DataFrame(mreg.params), pd.DataFrame(m.params)], axis=1)
    print(df)
    print(f"\npre.rsquared: {mreg.rsquared}")
    print(f"pre.rsquared_adj: {mreg.rsquared_adj}\n")
    print(f"post.rsquared: {m.rsquared}")
    print(f"post.rsquared_adj: {m.rsquared_adj}")
    print("====================================================================================================^")
```

```python
column_names = x_train.columns
for i in range(5):
    for j in range(5, 10):
        plot_interaction(x_train.iloc[:,i], x_train.iloc[:,j], (column_names[i], column_names[j]))
```

```python
interaction_cols = \
pd.concat([(watcher * free_return).rename('watcher * free_return'),
          (watcher * refurbished).rename('watcher * refurbished'),
          (shipping * benefits_charity).rename('shipping * benefits_charity'),
          (price * shipping).rename('price * shipping'),
          (num_ratings * shipping).rename('num_ratings * shipping')], axis = 1)

x_train = pd.concat([x_train, interaction_cols], axis = 1)

log_y = np.log(sold+1)
fit = smf.ols(formula = 'log_y ~ x_train', data = pd.concat([x_train, log_y], axis=1), missing = 'drop').fit()
fit.summary()
```
