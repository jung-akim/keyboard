### This file does following things:
- Remove the influential outlier
- Fit Linear Regression and compute NMAE and R2
- Fit Poisson Regression and compute NMAE and R2
- Cross-validation check
- Regularization
    - Lasso/Ridge for Linear Regression
    - Elastic Net for Poisson Regression
- Final Modeling
- Prediction plots
- Inferences from the model
- Removed outlier's info

```python
# Necessary imports
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.compat.python import lrange
import patsy as pt
from scipy import stats
from statsmodels.graphics.gofplots import ProbPlot
from statsmodels import graphics
from statsmodels.tools.eval_measures import rmse
from sklearn.metrics import mean_absolute_error
from statsmodels.tools.tools import add_constant
import pickle

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso, Ridge
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline

%matplotlib inline
```

- Read the feature-engineered training set and raw test set from feature_engineering.ipynb file
- Insert Missing value Indicator variables for test set: price, rating, shipping, 
status(of the product: either open_box, pre_owned, refurbished, brand_new(baseline))<br>
- Apply the same feature transformation on the test set
- Impute test data's missing values with -1
- Split the test set into validation and test set

```python
x_train = pd.read_pickle('data/x_train.pkl')
x_test = pd.read_pickle('data/x_test.pkl')
y_train = pd.read_pickle('data/y_train.pkl')
y_test = pd.read_pickle('data/y_test.pkl')

# Missing value Indicator variables for test set:
# price, rating, shipping, 
# status(of the product: either open_box, pre_owned, refurbished, brand_new(baseline))
x_test.loc[:,'price_present'] = (~x_test['price'].isnull()).astype(int)
x_test.loc[:,'rating_present'] = (~x_test['rating'].isnull()).astype(int)
x_test.loc[:,'shipping_present'] = (~x_test['shipping'].isnull()).astype(int)
x_test.loc[:,'status_present'] = (~x_test['open_box'].isnull()).astype(int)

# Feature engineer  main effects
x_test.loc[:,['price', 'num_ratings', 'shipping']] = \
pd.concat([1/np.sqrt(x_test.price + 1), 1/(x_test.num_ratings + 1), 1/np.sqrt(x_test.shipping + 1)], axis =1)


# Feature engineer  interaction effects
interaction_cols = \
pd.concat([(x_test.watcher * x_test.free_return).rename('watcher * free_return'),
          (x_test.watcher * x_test.refurbished).rename('watcher * refurbished'),
          (x_test.shipping * x_test.benefits_charity).rename('shipping * benefits_charity'),
          (x_test.price * x_test.shipping).rename('price * shipping'),
          (x_test.num_ratings * x_test.shipping).rename('num_ratings * shipping')], axis = 1)

x_test = pd.concat([x_test, interaction_cols], axis = 1)
x_test = x_test.fillna(-1) # Impute missing values with -1

# Re-split test data into test data and validation data
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=0.5, random_state = 1)
```

#### Remove influential outlier (Cook's D) from training set

```python
"""
Reference:
https://www.kaggle.com/mryder73/regression-challenge-day-2-python

get data relating to high leverage points using statsmodels
leverage info doesn't appear to be available for a poisson distribution
gaussian model used for illustration
"""

# fit OLS regression model 
model_g = sm.OLS(y_train, x_train).fit()

# leverage, from statsmodels
model_leverage = model_g.get_influence().hat_matrix_diag
# cook's distance, from statsmodels
model_cooks = model_g.get_influence().cooks_distance[0]

# plot cook's distance vs high leverage points
sns.regplot(model_leverage, model_cooks, fit_reg=False)
plt.xlim(xmin=-0.005, xmax=0.02)
plt.xlabel('Leverage')
plt.ylabel("Cook's distance")
plt.title("Cook's vs Leverage");
```

Remove the outlier from training set

```python
mask = ~(y_train == y_train.iloc[model_cooks.argmax()]) # Without influential outlier(Cook's D)
x_train, y_train = x_train.loc[mask], y_train[mask]
```

## Linear regression

```python
# Note: y = log(y + 1) and price, rating, num_ratings, and shipping are transformed
X, y = x_train, np.log(y_train+1)

formula = 'y ~ price + rating + num_ratings + watcher + shipping' + \
' + free_return + open_box + pre_owned + refurbished + benefits_charity' + \
' + price_present + rating_present + shipping_present + status_present' + \
' + price * free_return + watcher * free_return + watcher * refurbished + shipping * benefits_charity' + \
' + price * shipping + num_ratings * shipping'


model0 = smf.ols(formula = formula, data = pd.concat([X, y], axis=1), missing = 'drop').fit()
model0.summary()
```

```python
plt.hist(np.exp(model0.fittedvalues) - 1, bins = np.linspace(0, 250, 51), label = 'Fitted', color = 'tomato')
plt.hist(y_train, bins = np.linspace(0, 250, 51), alpha = 0.5, label = 'Actual', color = 'dodgerblue')
plt.xlim(0, 250)
plt.legend()
plt.title('Linear Regression')
plt.xlabel('Sale Volume in training set', fontsize = 20)
plt.ylabel('Frequency', fontsize = 20);
print(f"MAE without the influential outlier = {mean_absolute_error(y_train, np.exp(model0.fittedvalues)-1)}")
print(f'NMAE without the influential outlier = {mean_absolute_error(y_train, np.exp(model0.fittedvalues)-1) / np.mean(np.abs(y_train - np.mean(y_train)))}')
plt.savefig('figures/appendix/{}'.format("LinReg Training set"), dpi=200, transparent=True, bbox_inches = "tight")
```

## Poisson regression

Residual analysis:<br>
https://www.kaggle.com/mryder73/regression-challenge-day-2-python

```python
X, y = x_train, y_train
X = sm.add_constant(X, has_constant='add' )
model1 = sm.GLM(y, X, family = sm.families.Poisson()).fit(method = 'ncg')

resid = model1.resid_deviance.copy()
resid_std = stats.zscore(resid)

sns.regplot(y_train, resid_std, fit_reg=True, marker = '.')
plt.title('Residual plot on training set')
plt.xlabel('Sold')
plt.ylabel('Standardized residuals')
plt.savefig('figures/appendix/{}'.format("PoissonReg Dev_Residual Plot - Training"), dpi=200, transparent=True, bbox_inches = "tight");

graphics.gofplots.qqplot(resid_std, line='r', marker = '.')
plt.ylim(-3,3)
plt.savefig('figures/appendix/{}'.format("PoissonReg Normal Plot - Training"), dpi=200, transparent=True, bbox_inches = "tight");
```

```python
X = pd.concat([x_train.copy(), y_train.copy()], axis = 1)

# add poisson fitted values to dataframe
X['reg_fit'] = model1.fittedvalues

# sort dataframe by 'price'
X.sort_values('price', inplace=True)

# plot & add a regression line
fig, ax = plt.subplots(1, 3, figsize = (13, 3))
X.sort_values('price', inplace=True)
sns.regplot(X['price'], X['sold'], fit_reg=False, marker = '.', ax = ax[0])
ax[0].plot(X['price'], X['reg_fit'])

X.sort_values('shipping', inplace=True)
sns.regplot(X['shipping'], X['sold'], fit_reg=False, marker = '.', ax = ax[1])
ax[1].plot(X['shipping'], X['reg_fit'])

X.sort_values('num_ratings', inplace=True)
sns.regplot(X['num_ratings'], X['sold'], fit_reg=False, marker = '.', ax = ax[2])
ax[2].plot(X['num_ratings'], X['reg_fit']);

print(f"MAE: {mean_absolute_error(y_train, model1.fittedvalues)}")
print(f'NMAE: {mean_absolute_error(y_train, model1.fittedvalues) / np.mean(np.abs(y_train - np.mean(y_train)))}')
print("Pseudo Rsquared",1 - model1.deviance / model1.null_deviance)
model1.summary()
```

```python
plt.hist(model1.fittedvalues, bins = np.linspace(0, 250, 51), label = 'Fitted', color = 'tomato')
plt.hist(y_train, bins = np.linspace(0, 250, 51), alpha = 0.5, label = 'Actual', color = 'dodgerblue')
plt.xlim(0, 250)
plt.legend()
plt.title('Poisson Regression')
plt.xlabel('Sale Volume in training set', fontsize = 20)
plt.ylabel('Frequency', fontsize = 20);
plt.savefig('figures/appendix/{}'.format("PoissonReg Training set"), dpi=200, transparent=True, bbox_inches = "tight");
```

## Negative Binomial regression

```python
X, y = x_train, y_train
X = sm.add_constant(X, has_constant='add')
model2 = sm.GLM(y, X, family=sm.families.NegativeBinomial(alpha=1e-10)).fit(method='ncg', disp = False)
plt.hist(model2.fittedvalues, bins = np.linspace(0, 250, 51), label = 'Fitted')
plt.hist(y_train, bins = np.linspace(0, 250, 51), alpha = 0.5, label = 'Actual')
plt.xlim(0, 250)
plt.legend()
plt.title('NegativeBinom');
print(f"MAE: {mean_absolute_error(y_train, model2.fittedvalues)}")
print(f'NMAE: {mean_absolute_error(y_train, model2.fittedvalues) / np.mean(np.abs(y_train - np.mean(y_train)))}')
print("Pseudo Rsquared:",1 - model2.deviance / model2.null_deviance)
model2.summary()
```

## Cross-validation

Model0 (Linear regression), Model1 (Poisson regression), Model2 (Negative binomial regression with dispersion parameter alpha)

```python
logynewpred =  model0.predict(x_val)
y_pred = np.exp(logynewpred) - 1
print(f"Linear regression MAE: {round(mean_absolute_error(y_pred, y_val), 1)}, R2 score: {round(r2_score(y_val, y_pred),2)}")
X = pd.DataFrame(add_constant(x_val.to_numpy(), has_constant='add'))
print(f"Poisson regression MAE: {round(mean_absolute_error(model1.predict(X), y_val), 1)}")
print(f"Negative Binomial MAE: {round(mean_absolute_error(model2.predict(X), y_val), 1)}")
print(f"MAE with sample mean(training) is {round(np.mean(np.abs(y_val - np.mean(y_train))), 1)}")
```

## Regularization

### Linear regression

```python
alphalist = 10**(np.linspace(-6,0,300))
err_vec_val = np.zeros(len(alphalist))
err_vec_train = np.zeros(len(alphalist))

for i,curr_alpha in enumerate(alphalist):

    # note the use of a new sklearn utility: Pipeline to pack
    # multiple modeling steps into one fitting process 
    steps = [('standardize', StandardScaler()), 
             ('lasso', Lasso(alpha = curr_alpha))]

    pipe = Pipeline(steps)
    pipe.fit(x_train.values, np.log(y_train+1))
    
    val_set_pred = np.exp(pipe.predict(x_val.values)) - 1
    err_vec_val[i] = mean_absolute_error(y_val, val_set_pred)
    
#plot the curve of validation error as alpha changes
plt.plot(np.log10(alphalist), err_vec_val)
plt.xlabel('log10 scale');
print('Minimum error:',np.min(err_vec_val))
print('Argmin alpha:',alphalist[np.argmin(err_vec_val)])

steps = [('standardize', StandardScaler()), 
         ('lasso', Lasso(alpha = 0.01714406135851972))]

pipe = Pipeline(steps)
pipe.fit(x_train.values, np.log(y_train+1))

val_set_pred = np.exp(pipe.predict(x_val.values)) - 1
print('R2 score:',r2_score(y_val, val_set_pred))
print('MAE:',mean_absolute_error(y_val, val_set_pred))
```

```python
plt.hist(val_set_pred, bins = np.linspace(0, 250, 51), label = 'Fitted', color = 'tomato')
plt.hist(y_val, bins = np.linspace(0, 250, 51), alpha = 0.5, label = 'Actual', color = 'dodgerblue')
plt.xlim(0, 250)
plt.legend()
plt.title('Linear Regression Regularized - Lasso')
plt.xlabel('Sale Volume in validation set', fontsize = 20)
plt.ylabel('Frequency', fontsize = 20);
plt.savefig('figures/appendix/{}'.format("LinReg Validation set - Lasso"), dpi=200, transparent=True, bbox_inches = "tight")
```

```python
alphalist = 10**(np.linspace(-2,2,200))
err_vec_val = np.zeros(len(alphalist))
err_vec_train = np.zeros(len(alphalist))

for i,curr_alpha in enumerate(alphalist):

    # note the use of a new sklearn utility: Pipeline to pack
    # multiple modeling steps into one fitting process 
    steps = [('standardize', StandardScaler()), 
             ('ridge', Ridge(alpha = curr_alpha))]

    pipe = Pipeline(steps)
    pipe.fit(x_train.values, np.log(y_train+1))
    
    val_set_pred = np.exp(pipe.predict(x_val.values))-1
    err_vec_val[i] = mean_absolute_error(y_val, val_set_pred)
    
# plot the curve of validation error as alpha changes
plt.plot(np.log10(alphalist), err_vec_val)
plt.xlabel('log10 scale');
print('Minimum error:',np.min(err_vec_val))
print('Argmin alpha:', alphalist[np.argmin(err_vec_val)])

steps = [('standardize', StandardScaler()), 
         ('ridge', Ridge(alpha = 100.0))]

pipe = Pipeline(steps)
pipe.fit(x_train.values, np.log(y_train+1))

val_set_pred = np.exp(pipe.predict(x_val.values)) - 1
print('R2:',r2_score(y_val, val_set_pred))
print('MAE:', mean_absolute_error(y_val, val_set_pred))
```

```python
plt.hist(val_set_pred, bins = np.linspace(0, 250, 51), label = 'Fitted', color = 'tomato')
plt.hist(y_val, bins = np.linspace(0, 250, 51), alpha = 0.5, label = 'Actual', color = 'dodgerblue')
plt.xlim(0, 250)
plt.legend()
plt.title('Linear Regression Regularized - Ridge');
plt.xlabel('Sale Volume in validation set', fontsize = 20)
plt.ylabel('Frequency', fontsize = 20);
plt.savefig('figures/appendix/{}'.format("LinReg Validation set - Ridge"), dpi=200, transparent=True, bbox_inches = "tight")
```

### Poisson regression

```python
alphalist = 10**(np.linspace(-1,3,200))
err_vec_val = np.zeros(len(alphalist))
err_vec_train = np.zeros(len(alphalist))

for i,curr_alpha in enumerate(alphalist):
    model = sm.GLM(y_train, std_x_train, family = sm.families.Poisson()).fit_regularized(alpha = curr_alpha)
    val_set_pred = model.predict(std_x_val)
    err_vec_val[i] = mean_absolute_error(y_val, val_set_pred)
    
# plot the curve of validation error as alpha changes
plt.plot(np.log10(alphalist), err_vec_val)
plt.xlabel('log10 scale');

print(np.min(err_vec_val))
print(alphalist[np.argmin(err_vec_val)])

model = sm.GLM(y_train, std_x_train, family = sm.families.Poisson()).fit_regularized(alpha = alphalist[np.argmin(err_vec_val)])
mean_absolute_error(y_val, model.predict(std_x_val))
```

## Final modeling: Poisson regression 

- Merge training set with validation set and assign it to X
- Standardize X with the overall mean and sd and add 1 to std_X values and add one-vector column.
    - 1 is added to std_X values to prevent division by zero during regularization process
- Standardize x_test as well with the same mean and sd and add 1 to it as well.

```python
X = pd.concat([x_train, x_val])
y = pd.concat([y_train, y_val])

std = StandardScaler()
std.fit(X.values)

std_X = sm.add_constant(std.transform(X.values) + 1, has_constant='add') # 1 is added to std_X values to prevent division by zero during regularization process
std_x_test = sm.add_constant(std.transform(x_test.values) + 1, has_constant='add')
```

```python
model = sm.GLM(y, std_X, family = sm.families.Poisson()).fit_regularized(alpha = 59.41133984965034)
list(zip(x_train.columns,model.params[1:]))
```

NMAE

```python
print(f"NMAE: {mean_absolute_error(y_test, model.predict(std_x_test)) / np.mean(abs(y_test - np.mean(y)))}")
print(f"MAE for the model: {mean_absolute_error(y_test, model.predict(std_x_test))}")
print(f"MAE with the sample mean of train+val: {np.mean(abs(y_test - np.mean(y)))}")
```

### Parameters in terms of x_test variables (un-standardizing and subtracting 1)

exp(standardized x * params) is the predicted value

On that note, compute parameters for "unstandardized" and "subtracted-of-1" explanatory variable values (to use the format of x_test, basically)

```python
params = model.params[[1, 5, 6, 12]].values / np.std(X)[['price', 'shipping', 'free_return', 'rating_present']].values
params
```

```python
const_multiplier = np.exp(-sum(model.params[[1,5,6,12]].values * \
np.mean(X)[['price', 'shipping', 'free_return', 'rating_present']].values / \
np.std(X)[['price', 'shipping', 'free_return', 'rating_present']].values) + \
sum(model.params[[1,5,6,12]].values))
const_multiplier
```
